```python
#!/usr/bin/env python
# coding: utf-8
import gym
import tensorflow as tf
# import tensorflow.contrib.slim as slim
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from tensorflow import keras
import matplotlib as mpl
import random
import os

data_size = 140
s_size = 125
a_size = 66
reward_list = []
state_list = []


for i in range(data_size):
    string = 'training_data/reward_' + str(i) + '.txt'
    tmp = []
    f = open(string)
    line = f.readline()
    while line:
        seperate_lines = line.split()
        reward = float(seperate_lines[0])
        tmp.append(reward)
        line = f.readline()
    reward_list.append(tmp)
    
for i in range(data_size):
    string = 'training_data/state_' + str(i) + '.txt'
    tmp = []
    f = open(string)
    line = f.readline()
    while line:
        seperate_lines = line.split()
        state = float(seperate_lines[0])
        tmp.append(state)
        line = f.readline()
    state_list.append(tmp)
print(len(reward_list[0]))        
print(len(state_list[0]))
```

```python
best_range_config = []
for i in range(data_size):
    maxx = 0
    for j in range(a_size):
        if(maxx < reward_list[i][j] ):
            maxx = reward_list[i][j]
    tmp_list = []        
    for j in range(a_size):
        if(maxx - 10 < reward_list[i][j] ):
            tmp_list.append(4)
        elif(maxx - 10 < reward_list[i][j]):
            tmp_list.append(3)
        elif(maxx - 20 < reward_list[i][j]):
            tmp_list.append(2)
        elif(maxx - 20 < reward_list[i][j]):
            tmp_list.append(1)
        else:
            tmp_list.append(0)
    best_range_config.append(tmp_list)
    
best_range_2Dconfig = []
for i in range(a_size):
    tmp = []
    for j in range(data_size):
        tmp.append(reward_list[j][i])
    best_range_2Dconfig.append(tmp)
    
    
# best_config = np.reshape(best_range_config, data_size)
```

```python
from sklearn.model_selection import train_test_split

state = np.reshape(state_list, [data_size, s_size])
state = (state/10000)
target = np.reshape(best_range_config, [data_size, a_size])
target2D = np.reshape(best_range_2Dconfig, [a_size, data_size])
target2D = (target2D)
target2D = target2D.T

X_train, X_test, y_train, y_test = train_test_split(
    state, target2D, test_size=0.1, random_state=42)
```

```python
y_train = y_train.T
y_test = y_test.T
```

```python

total_episodes = 500
model_arr = []

for i in range(a_size):
    model = keras.Sequential()
    model.add(keras.layers.Dense(s_size, input_shape=(s_size,), activation='relu'))
    model.add(keras.layers.Dense(64, activation='relu'))
    model.add(keras.layers.Dropout(0.2))
    model.add(keras.layers.Dense(32, activation='relu'))
    model.add(keras.layers.Dense(1, activation='relu'))
    model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
                  loss='mse')
    model_arr.append(model)

for e in range(total_episodes):
    for i in range(a_size):
        print('model ' + str(i) + ' index')
        model_arr[i].fit(X_train, y_train[i], verbose=1)
```

```python
test_number = 8
next_state = np.reshape(X_test[test_number], [1, s_size])
y_pred = []
y_gd = []
for i in range(a_size):
    y_pred.append((model_arr[i].predict(next_state)[0]))
    y_gd.append(y_test[i][test_number])
print(y_pred)
print(y_gd)
```

```python

print("Plot Learning Performance")
mpl.rcdefaults()
mpl.rcParams.update({'font.size': 16})

fig, ax = plt.subplots(figsize=(10,4))
plt.grid(True, linestyle='--')
plt.title('Learning Performance')
plt.plot(range(len(y_gd)), y_gd, label='gd', marker="^", linestyle=":")#, color='red')
plt.plot(range(len(y_pred)), y_pred, label='predict', marker="*", linestyle="-")#, color='k')
plt.xlabel('Episode')
plt.ylabel('packet loss')
plt.legend(prop={'size': 12})

plt.savefig('learning.pdf', bbox_inches='tight')
plt.show()
```
