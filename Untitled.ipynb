```python
#!/usr/bin/env python
# coding: utf-8
import gym
import tensorflow as tf
# import tensorflow.contrib.slim as slim
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from tensorflow import keras
import matplotlib as mpl
import random
import os
from os import listdir
from os.path import isfile, isdir, join

data_size = 500
s_size = 125
a_size = 9
reward_list = []
state_list = []
base_data_dir = 'training_data/'
# base_data_dir = 'training_data/125_uniform_traffic_data_1_10000/'

files = listdir(base_data_dir)

for f in files:
    fullpath = join(base_data_dir, f)

    if isfile(fullpath):
        continue
    if (str(fullpath) == 'training_data/.ipynb_checkpoints'):
        continue
    elif isdir(fullpath):
        print("目錄：", f)
#         fullpath = join(fullpath, f)

        for i in range(data_size):
            string = str(fullpath) + '/reward_' + str(i) + '.txt'
            tmp = []
            f = open(string)
            line = f.readline()
            while line:
                seperate_lines = line.split()
                reward = float(seperate_lines[8])
                tmp.append(reward)
                line = f.readline()
            reward_list.append(tmp)

        for i in range(data_size):
            string = str(fullpath) + '/state_' + str(i) + '.txt'
            tmp = []
            f = open(string)
            line = f.readline()
            while line:
                seperate_lines = line.split()
                state = float(seperate_lines[0])
                tmp.append(state)
                line = f.readline()
            state_list.append(tmp)
print(len(reward_list))        
print(len(state_list))
```

```python
best_range_config = []
for i in range(data_size):
    maxx = 0
    for j in range(a_size):
        if(maxx < reward_list[i][j] ):
            maxx = reward_list[i][j]
    tmp_list = []        
    for j in range(a_size):
        if(maxx - 5 < reward_list[i][j] ):
            tmp_list.append(5)
        elif(maxx - 10 < reward_list[i][j]):
            tmp_list.append(4)
        elif(maxx - 15 < reward_list[i][j]):
            tmp_list.append(3)
        elif(maxx - 20 < reward_list[i][j]):
            tmp_list.append(2)
        else:
            tmp_list.append(1)
    best_range_config.append(tmp_list)
    
best_range_2Dconfig = []
for i in range(a_size):
    tmp = []
    for j in range(len(reward_list)):
        tmp.append(reward_list[j][i])
    best_range_2Dconfig.append(tmp)
    
    
# best_config = np.reshape(best_range_config, data_size)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from scipy.sparse import coo_matrix

from sklearn.utils import shuffle


state = np.reshape(state_list, [len(state_list), len(state_list[0])])
state = ((state - np.min(state))/(np.max(state) - np.min(state)))
# target = np.reshape(best_range_config, [data_size*2, a_size])
target2D = np.reshape(best_range_2Dconfig, [a_size, len(state_list)])
target2D = (target2D)
target2D = target2D.T

pca = PCA(n_components=80) # n_components can be integer or float in (0,1)
pca.fit(state)  # fit the model
print(pca.fit_transform(state))  # transformed data
state = pca.fit_transform(state)

X_train, X_test, y_train, y_test = train_test_split(
    state, target2D, test_size=0.1, random_state=42)



# X_sparse = coo_matrix(X_train)
# X_train, X_sparse, y_train = shuffle(X_train, X_sparse, y_train, random_state=0)
```

```python
y_train = y_train.T
y_test = y_test.T
```

```python

total_episodes = 500
input_size = 80
model_arr = []

for i in range(a_size):
    model = keras.Sequential()
    model.add(keras.layers.Dense(input_size, input_shape=(input_size,), activation='relu'))
    model.add(keras.layers.Dense(64, activation='relu'))
    model.add(keras.layers.Dropout(0.01))
    model.add(keras.layers.Dense(16, activation='relu'))
    model.add(keras.layers.Dense(1, activation='relu'))
    model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
                  loss='mse')
    model_arr.append(model)

for e in range(total_episodes):
    for i in range(a_size):
        print('model ' + str(i) + ' index')
        model_arr[i].fit(X_train, y_train[i], verbose=1)
```

```python
test_state_number = 50
test_config_vumber = 1
next_state = np.reshape(X_test[test_state_number], [1, input_size])
y_pred = []
y_gd = []
for i in range(a_size):
    y_pred.append((model_arr[i].predict(next_state)[0]))
    y_gd.append(y_test[i][test_state_number])
# for i in range(len(X_test)):
#     next_state = np.reshape(X_test[i], [1, s_size])
#     y_pred.append((model_arr[test_config_vumber].predict(next_state)[0]))
#     y_gd.append(y_test[test_config_vumber][i])
print(y_pred)
print(y_gd)
```
