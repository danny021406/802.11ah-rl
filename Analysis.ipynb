```python
import gym
import tensorflow as tf
# import tensorflow.contrib.slim as slim
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from tensorflow import keras
import matplotlib as mpl
import random
import os
import math

f = open("training_info.txt")
rew_history = []
err_history = []

line = f.readline()
while line:
    seperate_lines = line.split()
    tmp = math.sqrt(math.pow( float(seperate_lines[1]), 2 ))
    if tmp > 100:
        tmp = 100
    err_history.append(tmp)
    line = f.readline()
err_history = err_history[:]
print("Plot Learning Performance")
mpl.rcdefaults()
mpl.rcParams.update({'font.size': 16})

fig, ax = plt.subplots(figsize=(10,4))
plt.grid(True, linestyle='--')
plt.title('Learning Performance')
plt.plot(range(len(rew_history)), rew_history, marker="^", linestyle=":")#, color='red')
plt.plot(range(len(err_history)), err_history, label='error', marker="^", linestyle=":")#, color='red')
plt.xlabel('step')
plt.ylabel('error')
plt.legend(prop={'size': 12})

plt.savefig('learning.pdf', bbox_inches='tight')
plt.show()
```
